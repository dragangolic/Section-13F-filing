## Enreaching data frame with more columns

```python
# Generating random values for price column
random_values = np.random.uniform(0, 350, size=len(df)).astype(float)
random_values = np.round(random_values, 2)
df['price'] = random_values
```

```python
df.head()
```

```python
# Generating random values for shares column
s = np.random.uniform(0,15000, size=len(df)).astype(int)
df['shares'] = s
```

```python
df.head()
```

```python
# Generate random integers (0 or 1) for mapping to 'EUR' and 'USD' for currency column
random_integers = np.random.randint(0, 2, size=len(df))

# Map the random integers to 'EUR' and 'USD'
currency_mapping = {0: 'EUR', 1: 'USD'}
random_currencies = [currency_mapping[i] for i in random_integers]
df['currency'] = random_currencies
```

```python
df.head()
```

```python
# Market Valuation column is product of price and shares columns
df['market valuation'] = df['price'] * df['shares']
```

```python
df.head()
```
